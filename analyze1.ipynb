```python
%matplotlib notebook
import matplotlib.pyplot as plt

# C:\Users\Михаил\Documents\GitHub\CellularAutomaton\1.txt
with open(r'C:\Users\Михаил\Documents\GitHub\CellularAutomaton\1.txt', 'r') as f:
    values = f.read().splitlines()
    #print(values)
    alive_and_dead = [i.split('/') for i in values]
    #print(alive_and_dead)
    alive_cells = [int(i[0]) for i in alive_and_dead]
    dead_cells = [int(i[1]) for i in alive_and_dead]
    ratio_cells = [int(i[0]) / (int(i[0])+int(i[1])) for i in alive_and_dead]
    print(ratio_cells)
    indexes = [i for i in range(len(values))]
plt.plot(indexes, alive_cells)
plt.plot(indexes, dead_cells)
plt.plot(indexes, ratio_cells)
plt.show()
```
